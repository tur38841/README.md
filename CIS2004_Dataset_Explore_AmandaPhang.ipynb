```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots()

ages = df.groupby("age").size()

x = ages.index
y = ages.values


ax.bar(x, y)
ax.set_xlabel("Age")
ax.set_ylabel("Number of People")
ax.set_title("Age vs Count")
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots()

ages = df.groupby("study_hours").size()

x = ages.index
y = ages.values


ax.bar(x, y)
ax.set_xlabel("Number of Studied Hours")
ax.set_ylabel("Number of People")
ax.set_title("Study vs Count")
plt.show()
plt.show()
```
